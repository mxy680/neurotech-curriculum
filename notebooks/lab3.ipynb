### 1.3. Creating Basic Plots

Let's start by creating some basic plots to get familiar with Matplotlib's functionality.

#### 1.3.1. Line Plot

Line plots are useful for showing trends over time or continuous data.


```python
# Line plot example
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

#### 1.3.2. Scatter Plot

Scatter plots are used to visualize the relationship between two variables.


```python
# Scatter plot example
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='r', s=100, alpha=0.7)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

#### 1.3.3. Bar Plot

Bar plots are useful for comparing discrete quantities across categories.


```python
# Bar plot example
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]

plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='c')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True, axis='y')
plt.show()
```

#### 1.3.4. Histogram

Histograms are used to represent the distribution of a dataset.


```python
import numpy as np

# Histogram example
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='m', edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True, axis='y')
plt.show()
```

## 2. Visualizing Neuroscience Data

Effective visualization is particularly important when working with complex neuroscience data, such as EEG signals and MRI scans. Matplotlib can help us gain insights from these types of data.

### 2.1. Plotting EEG Signals

EEG (electroencephalography) signals are often visualized as time-series data. Let's create a plot for a simulated EEG signal.



```python
# Simulated EEG signal
time = np.linspace(0, 10, 500)
eeg_signal = np.sin(2 * np.pi * 1 * time) + 0.5 * np.random.randn(500)

plt.figure(figsize=(12, 6))
plt.plot(time, eeg_signal, color='b')
plt.title('Simulated EEG Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

### 2.2. Visualizing MRI Scans

MRI scans are typically visualized as images. We can use Matplotlib to display MRI images and overlay various types of analysis.

#### 2.2.1. Displaying an MRI Image

Let's simulate displaying an MRI image using a sample image array.


```python
import matplotlib.image as mpimg

# Simulate an MRI image
mri_image = np.random.rand(100, 100)  # Simulated MRI image data

plt.figure(figsize=(8, 8))
plt.imshow(mri_image, cmap='gray', origin='upper')
plt.title('Simulated MRI Image')
plt.colorbar()
plt.show()
```

### 2.3. Overlaying Analysis on MRI Images

You can overlay additional analysis on MRI images, such as highlighting areas of interest.

#### 2.3.1. Adding Annotations

We can annotate specific regions or add markers to the MRI image.


```python
# Overlay analysis on MRI image
plt.figure(figsize=(8, 8))
plt.imshow(mri_image, cmap='gray', origin='upper')
plt.title('MRI Image with Annotations')

# Add annotations
plt.scatter([30, 70], [30, 70], color='red', s=100, marker='o', label='Areas of Interest')
plt.legend()

plt.colorbar()
plt.show()
```

## 3. Advanced Visualization Techniques

### 3.1. Subplots

Subplots allow you to create multiple plots in a single figure. This is useful for comparing different visualizations.



```python
# Subplots example
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Line Plot')

# Scatter plot
axs[0, 1].scatter(x, y, color='r', s=100, alpha=0.7)
axs[0, 1].set_title('Scatter Plot')

# Bar plot
axs[1, 0].bar(categories, values, color='c')
axs[1, 0].set_title('Bar Plot')

# Histogram
axs[1, 1].hist(data, bins=30, color='m', edgecolor='black')
axs[1, 1].set_title('Histogram')

for ax in axs.flat:
    ax.label_outer()  # Hide inner labels

plt.tight_layout()
plt.show()
```

### 3.2. Customizing Plots

Matplotlib provides extensive options for customizing plots, including colors, labels, and styles.



```python
# Customizing plots
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='g', linewidth=2, markersize=10)
plt.title('Customized Line Plot', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)
plt.grid(True)
plt.show()
```
